```python
import os
os.environ["KERAS_BACKEND"] = "theano"
import keras
import numpy as np
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
(Train_data, Train_label), (Test_data, Test_label) = keras.datasets.imdb.load_data(num_words=2000)
```

```python
print("Train_data : {}".format(Train_data))
print("Train_data shape : {}".format(Train_data.shape))
print("Train_label: {}".format(Train_label))
print("Train_label shape : {}".format(Train_label.shape))
print("Test_data : {}".format(Test_data))
print("Test_data shape : {}".format(Test_data.shape))
print("Test_label: {}".format(Test_label))
print("Test_label shape : {}".format(Test_label.shape))
```

```python
max_words = 2000
max_length = 100
Train_data = keras.preprocessing.sequence.pad_sequences(Train_data, maxlen=max_length)
Test_data = keras.preprocessing.sequence.pad_sequences(Test_data, maxlen=max_length)
```

```python
print(Train_data[1])
print(Train_label[1])
print(Train_data.shape)
print(Train_label.shape)
```

```python
model = keras.models.Sequential([
    keras.layers.Embedding(max_words, 128),
    keras.layers.Dropout(0.1),
    keras.layers.LSTM(64),
    keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(loss="binary_crossentropy",metrics=["accuracy"],optimizer="adam")
```

```python
model.fit(Train_data,Train_label,epochs=2,batch_size=32)
```

```python
prediction = model.predict(Test_data)
```

```python
print(np.round(prediction[0]))
print(Test_label[0])
```
